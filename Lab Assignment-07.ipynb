```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('/content/world_population.csv')

# Plotting the line chart
plt.figure(figsize=(10,6))
plt.plot(data['Year'], data['Population'], marker='o', color='blue', linestyle='-', linewidth=2)

# Add labels and title
plt.xlabel('Year', fontsize=12)
plt.ylabel('Population (in thousands)', fontsize=12)
plt.title('World Population Growth Over Years', fontsize=14)
plt.grid(True)

# Display the plot
plt.show()
```

```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.dates import DateFormatter

# Load dataset
data = pd.read_csv('/content/AAPL.csv', parse_dates=['Date'])

# Sort by date (if not already sorted)
data.sort_values('Date', inplace=True)

# Plotting the line chart
plt.figure(figsize=(12,6))
plt.plot(data['Date'], data['Close'], color='blue', linewidth=2, label='Close Price')

# Format x-axis for dates
plt.gcf().autofmt_xdate()  # Auto rotates dates
date_format = DateFormatter("%Y-%m")
plt.gca().xaxis.set_major_formatter(date_format)

# Labels and title
plt.xlabel('Date', fontsize=12)
plt.ylabel('Closing Price (USD)', fontsize=12)
plt.title('Apple (AAPL) Stock Prices Over Time', fontsize=14)
plt.legend()
plt.grid(True)

# Show plot
plt.show()

```
